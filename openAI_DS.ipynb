##Automobile assist Application - using Gradio UI

Using Gradio UI for chat interface - An application where an Automated Agent assists customers on what car to buy! 

Using OpenAI

```python
import os
import openai
import gradio as gr

#code to get apikey from file but have hardcoded for now
#def open_file(filepath):
#    with open(filepath, 'r', encoding='utf-8') as infile:
#        return infile.read()

#removing api key due to security reasons before posting in github
openai.api_key = ""

#open_file('/content/sample_data/openai_apikey.txt')

start_sequence = "\n Agent:"
restart_sequence = "\n Customer:"

prompt = "The following is a conversation between a customer who wants to buy a car and an agent who assists the customer and guides him on what car to by based on his or her needs. The agent is helpful, creative, clever, and very friendly.\n\nCustomer: Hi, I am looking to buy a car. I want someone to guide me on what car would best suit my needs.\nAgent: Sure I can help you with that. \nCustomer: "

def openai_create(prompt):
    response = openai.Completion.create(
    model="text-davinci-003",
    prompt=prompt,
    temperature=0.9,
    max_tokens=150,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0.6,
    stop=[" Customer:", " Agent:"]
    )
    return response.choices[0].text


#the input is text which is given to gradio app
#history is to store the state of the application - to keep the context of past 
#appending every input and output to history and it is stored in prompt
def chatgpt_clone(input, history):
    history = history or []
    s = list(sum(history, ()))
    s.append(input)
    inp = ' '.join(s)
    output = openai_create(inp)
    history.append((input, output))
    return history, history

#using blocks to create gradio application
block = gr.Blocks()


with block:
    #title
    gr.Markdown("""<h3><center>Automated Agent to assist customers on what car to buy!</center></h3>
    """)
    chatbot = gr.Chatbot()
    message = gr.Textbox(placeholder="Type Here")
    state = gr.State()
    submit = gr.Button("Submit")
    submit.click(chatgpt_clone, inputs=[message, state], outputs=[chatbot, state])

block.launch(debug = True)
```

##Here we are creating a model, fine-tuning the model and using the model in an application

- The model is given a prompt about a year, a place(country) and 2 sectors on which scenarios are generated, the model is going to be trained and the model would predict about. 

- Inorder to fine tune, here we have a 4 parameters, and 4 values for each parameter, the loop runs in every combination of the parameter(4 * 4 * 4 * 4 = 256) times, hence 256 sequence of text is generated and trained in the model. This helps finetune the model.

- Each loop, the text generated is stored in a file, which is later added to a file in JSON format.


Designed 1 application which worked, couldn't try another one as I ran out of free trial with open AI

```python
import openai
from time import time,sleep
from uuid import uuid4

def open_file(filepath):
  with open(filepath, 'r', encoding='utf-8') as infile:
    return infile.read()


def save_file(filepath, content):
  with open(filepath, 'w', encoding='utf-8') as outfile:
    outfile.write(content)

#open.api_key = open_file('openapikey.txt')
#removing api key due to security reasons before posting in github
openai.api_key = ""

places = ['America', 'India', 'China', 'UK']

years = ['2023', '2026', '2029', '2032']

para1s = ['environment','world','society','people']

para2s = ['job', 'transportation', 'technology', 'science']

def gpt3_completion(prompt, engine='text-davinci-002', temp=1.0, top_p=1.0, tokens=1000, freq_pen=0.0, pres_pen=0.0, stop=['asfgfdh']):
  max_try = 5
  retry = 0
  prompt = prompt.encode(encoding='ASCII',errors='ignore').decode()

  while True:
    response = openai.Completion.create(engine=engine,prompt=prompt,temperature=temp,max_tokens=tokens,top_p=top_p,frequency_penalty=freq_pen,presence_penalty=pres_pen,stop=stop)
    text = response['choices'][0]['text'].strip()
    filename = '%s_gpt3.txt' % time()
    print('/content/sample_data/gpt3_logs/%s' % filename, prompt + '\n\n==========\n\n' + text)
    save_file('/content/sample_data/gpt3_logs/%s' % filename, prompt + '\n\n==========\n\n' + text)
    return text
  
if __name__ == '__main__':
  count = 0
  for place in places:
    for year in years:
      for para1 in para1s:
        for para2 in para2s:
          count += 1
          prompt = open_file('/content/sample_data/finetune_prompt.txt')
          prompt = prompt.replace('<<place>>', place)
          prompt = prompt.replace('<<year>>', year)
          prompt = prompt.replace('<<para1>>', para1)
          prompt = prompt.replace('<<para2>>', para2)
          prompt = prompt.replace('<<UUID>>', str(uuid4()))
          print('\n\n', prompt)
          completion = gpt3_completion(prompt)
          outprompt = 'year : %s\nplace: %s\nparameter1 : %s\nparameter2 : %s\n\nScenario:' %(year, place, para1, para2)
          filename = (place+year+para1+para2).replace(' ','').replace('&','') + '%s.txt' %time()
          save_file('/content/sample_data/prompts/%s' %filename, outprompt)
          save_file('/content/sample_data/completion/%s' %filename, completion)
          print('\n\n', outprompt)
          print('\n\n', completion)
          if count >500:
            exit()			

```

All the scenarios text generated is added to a file in JSON format

```python
import os
import json

src_dir = '/content/sample_data/completion/'
prompt_dir = '/content/sample_data/prompts/'

def open_file(filepath):
  with open(filepath, 'r', encoding='utf-8') as infile:
    return infile.read()

if __name__ == '__main__':
  files = os.listdir(src_dir)
  data=list()
  for file in files:
    completion = open_file(src_dir+file)
    prompt = open_file(prompt_dir+file)
    info = {'prompt':prompt, 'completion':completion}
    data.append(info)
  with open('/content/sample_data/agi_scenarios.jsonl', 'w') as outfile:
    for i in data:
      json.dump(i, outfile)
      outfile.write('\n')

```

From the JSON file created with the scenario prompt texts, a model is created in OpenAI. This model can then be used to predict various scenarios.   

```python
import requests
import openai
from pprint import pprint

#removing api key due to security reasons before posting in github
openai.api_key = ""


def file_upload(filename, purpose="fine-tune"):
  resp = openai.File.create(purpose=purpose, file=open(filename))
  pprint(resp)
  return resp

def file_list():
  resp = openai.File.list()
  pprint(resp)

def finetune_model(fileid, suffix, model='davinci'):
  header={'Content-Type': 'application/json', 'Authorization': 'Bearer %s' %openai.api_key}
  payload= {'training_file': fileid, 'model':model, 'suffix':suffix}
  resp= requests.request(method='POST', url = 'https://api.openai.com/v1/fine-tunes', json=payload, headers = header, timeout=45)
  pprint(resp.json())

def finetune_list():
  header = {'Content-Type': 'application/json', 'Authorization': 'Bearer %s' %openai.api_key}
  resp = requests.request(method='GET', url = 'https://api.openai.com/v1/fine-tunes', headers=header, timeout = 45)
  pprint(resp.json())

def finetune_events(ftid):
  header = {'Content-Type': 'application/json', 'Authorization': 'Bearer %s' %openai.api_key}
  resp = requests.request(method='GET', url = 'https://api.openai.com/v1/fine-tunes/%s/events'%ftid, headers=header, timeout = 45)
  pprint(resp.json())

def finetune_get(ftid):
  header = {'Content-Type': 'application/json', 'Authorization': 'Bearer %s' %openai.api_key}
  resp= requests.request(method='GET', url = 'https://api.openai.com/v1/fine-tunes/%s' %ftid, headers = header, timeout=45)
  pprint(resp.json())

resp = file_upload('/content/sample_data/agi_scenarios.jsonl')
finetune_model(resp['id'], 'agi_scenarios_model', 'davinci')
finetune_list()
```
